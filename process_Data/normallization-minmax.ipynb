### Train data

```python
# create a sample data frame
df_train = pd.read_csv(TRAIN_PATH)
df_test_csic = pd.read_csv(TEST_csic2010_PATH)
df_test_fwaf = pd.read_csv(TEST_FWAF_PATH)
df_test_http = pd.read_csv(TEST_https_PATH)
```

```python
cols_to_keep = df_train.columns[:-1]
last_column = df_train.iloc[:, -1] 
# create a new dataframe with selected columns
df_new = df_train[cols_to_keep]
scaler = MinMaxScaler()
train_file = pd.DataFrame(scaler.fit_transform(df_new), columns=df_new.columns)
```

```python
# print the last column
# print(last_column)

# # get the last column as a dataframe
# last_column_df = df_train.iloc[:, [-1]]

# # print the last column as a dataframe
# print(last_column_df)
```

```python
train_file['Label'] = last_column
# display(train_file)
```

```python
train_file.to_csv(TRAIN_PATH_minmax,index=False) 
```
