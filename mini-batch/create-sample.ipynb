```python

# Define the dataset (assuming it's a numpy array)
# dataset = np.array([...])  # Replace [...] with your dataset

# Define the parameters
n = 32  # Number of consecutive samples in each mini-batch

# Randomly permute the dataset
np.random.shuffle(dataset)

# Generate a mini-batch
def generate_mini_batch(dataset, n):
    # Choose a random starting index
    start_index = np.random.randint(0, len(dataset) - n - 1)
    
    # Select n+1 consecutive samples starting from the random index
    mini_batch = dataset[start_index : start_index + n + 1]
    
    return mini_batch

# Generate a mini-batch using the defined structure
mini_batch = generate_mini_batch(dataset, n)

# Print the mini-batch
print(mini_batch)
```
