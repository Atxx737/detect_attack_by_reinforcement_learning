# CREATE dataset


### data from OK_csic2010

```python
# get good data
df_good = pd.read_csv(GOOD_OK_csic2010_src)
good_to_test = df_good.sample(frac=0.2, random_state=1)
good_to_test.shape[0]
```

```python
# get bad data
df_bad = pd.read_csv(BAD_OK_csic2010_src)
bad_to_test = df_bad.sample(frac=0.2, random_state=1)
bad_to_test.shape[0]
```

```python
# create test from csic2010
TEST_csic2010 =  pd.concat([good_to_test,bad_to_test]) 
TEST_csic2010.shape[0]
TEST_csic2010.to_csv(TEST_OK_csic2010,index=False) 
```

```python
# get 80% data to train
good_to_train = df_good.drop(good_to_test.index)
bad_to_train = df_bad.drop(bad_to_test.index)
TRAIN_csic2010 = pd.concat([good_to_train,bad_to_train]) 
TRAIN_csic2010.shape[0]
```

### data from OK_fwaf

```python
# get good data
df_good = pd.read_csv(GOOD_OK_fwaf_queries_src)
good_to_test = df_good.sample(frac=0.2, random_state=1)
good_to_test.shape[0]
```

```python
#get bad data
df_bad = pd.read_csv(BAD_OK_fwaf_queries_src)
bad_to_test = df_bad.sample(frac=0.2, random_state=1)
bad_to_test.shape[0]
```

```python
# create test from fwaf
TEST_fwaf =  pd.concat([good_to_test,bad_to_test]) 
TEST_fwaf.shape[0]
TEST_fwaf.to_csv(TEST_OK_fwaf_queries,index=False) 
```

```python
# get 80% to train
good_to_train = df_good.drop(good_to_test.index)
bad_to_train = df_bad.drop(bad_to_test.index)
TRAIN_fwaf = pd.concat([good_to_train,bad_to_train])
TRAIN_fwaf.shape[0]
```

### data from http-params

```python
# get good data
df_good = pd.read_csv(GOOD_OK_http_param_src)
good_to_test = df_good.sample(frac=0.2, random_state=1)
good_to_test.shape[0]
```

```python
#get bad data
df_bad = pd.read_csv(BAD_OK_http_param_src)
bad_to_test = df_bad.sample(frac=0.2, random_state=1)
bad_to_test.shape[0]
```

```python
# create test from http
TEST_http =  pd.concat([good_to_test,bad_to_test]) 
TEST_http.shape[0]
TEST_http.to_csv(TEST_OK_http_param,index=False) 
```

```python
# get 80% to train
good_to_train = df_good.drop(good_to_test.index)
bad_to_train = df_bad.drop(bad_to_test.index)
TRAIN_http_param = pd.concat([good_to_train,bad_to_train])
TRAIN_http_param.shape[0]
```

### train data

```python
connect_data = pd.concat([TRAIN_csic2010,TRAIN_fwaf,TRAIN_http_param])

shuffle(connect_data).to_csv(TRAIN_PATH,index=False) 
```
